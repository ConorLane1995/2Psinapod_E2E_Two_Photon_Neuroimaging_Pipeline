```python
def get_mean_and_SE(percentages):

    means = np.mean(percentages, axis=0)
    
    SE = scipy.stats.sem(percentages, axis=0)

    return means, SE
```

```python
from statsmodels.stats.multitest import multipletests


def get_corrected_p_values(data):

    p_values = []
    W_stats = []

    for i in range(len(data[0,:])):

        statistic,p_value = stats.wilcoxon(data[:,i],alternative='less')
        p_values.append(p_value)
        W_stats.append(statistic)


    corrected_p_values = multipletests(p_values, method='fdr_bh')[1]



    return W_stats,corrected_p_values

W_stats,p_values = get_corrected_p_values(max_percent_diffs_day_1_6_saline)
print(p_values)
print(W_stats)

```

```python
from statsmodels.stats.multitest import multipletests


def get_corrected_p_values_t_tests(data1,data2):

    p_values = []
    t_stats = []

    for i in range(len(data1[0,:])):

        statistic,p_value = stats.mannwhitneyu(data1[:,i],data2[:,i])
        p_values.append(p_value)
        t_stats.append(statistic)


    corrected_p_values = multipletests(p_values, method='fdr_bh')[1]



    return t_stats,corrected_p_values

t_stats,p_values = get_corrected_p_values_t_tests(max_percent_diffs_day_1_21_saline,max_percent_diffs_day_1_21_psilocybin)
print(p_values)
print(t_stats)
```
